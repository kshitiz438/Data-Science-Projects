## Data Preprocessing

```python
# First we need to convert categorical variables into Numeric in order to find the most informative set of Feature
# to improve the performance of the classifier.

from sklearn.preprocessing import LabelEncoder
enc = {}
for col in airline.select_dtypes(include = ['object']).columns:
    enc[col] = LabelEncoder()
    airline[col] = enc[col].fit_transform(airline[col])
    
airline.head()
```

```python
# We have successfully converted Categorical values into numeric values
# Now we use a Heatmap to find the correlation between the features

corr = airline.corr()
plt.figure(figsize = (40, 25))
sns.heatmap(corr, annot = True, linewidth = 0.5, cmap = 'YlGnBu')
plt.show()
```

```python
from sklearn.model_selection import train_test_split #Import train_test_split function
from sklearn import metrics #Import scikit-learn mertics function for precision, accuracy, confusion matrix calculations
X = airline.drop('satisfaction', axis = 1) # Collecting the independent variables
y = airline['satisfaction'] # Collecting the dependent Variable
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

## Decision Tree Classifier

```python
# Our First model is Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc = dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

```python
# Evaluating our Model

print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print("Precision: ", metrics.precision_score(y_test, y_pred))
print("Recall: ", metrics.recall_score(y_test, y_pred))
print("F1 score: ", metrics.f1_score(y_test, y_pred))
```

```python
# Let's print the confusion matrix

cm = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4, 2))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', square=True, cbar=False, linewidth = 0.5)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

## Random Forest Classifier

```python
# Our second model is Random forest classifier

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf = rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
```

```python
# Evaluating our Model

print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print("Precision: ", metrics.precision_score(y_test, y_pred))
print("Recall: ", metrics.recall_score(y_test, y_pred))
print("F1 score: ", metrics.f1_score(y_test, y_pred))
```

## Compairing our Models

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier

from sklearn.model_selection import cross_val_score

predictors = ['Gender', 'Customer Type', 'Age', 'Type of Travel', 'Class', 'Flight Distance', 'Inflight wifi service', 
              'Departure/Arrival time convenient', 'Ease of Online booking', 'Gate location', 'Food and drink', 
              'Online boarding', 'Seat comfort', 'Inflight entertainment', 'On-board service', 'Leg room service', 
              'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness', 'Departure Delay in Minutes', 
              'Arrival Delay in Minutes']

models = []

models.append(("Logistic Regression: ", LogisticRegression()))
models.append(("Decision Tree: ", DecisionTreeClassifier()))
models.append(("Random Forest: ", RandomForestClassifier()))
models.append(("Naive bayes: ", GaussianNB()))
models.append(("Gradient Boosting Classifier: ", GradientBoostingClassifier()))


results = []
names = []
for name,model in models:
    result = cross_val_score(model, airline[predictors], airline["satisfaction"],  cv=3)
    names.append(name)
    results.append(result)

for i in range(len(names)):
    print(names[i],results[i].mean())
```

```python
""" Here we can see that Random Forest Classifier wins as it has given us the most accurate result """
```
