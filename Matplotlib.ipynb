```python
ypoints = np.array([4, 15, 8, 20, 11, 15, 8])

plt.plot(ypoints)
plt.show
```

```python
plt.plot(ypoints)
plt.ylabel('Samples')
plt.xlabel('Values')
plt.title('Figure 1')
plt.show()
```

```python
plt.plot(ypoints, marker = 'o', ms = 5)
plt.show()
```

```python
plt.plot(ypoints, '*--y')         #  "--" means dotted lines, r means line colour, * means marker shape
plt.show()
```

```python
plt.plot(ypoints, '*g', ms = 20, mec = 'r')        #mec = Marker Edge Color
plt.show()
```

```python
plt.plot(ypoints, '-*', ms = 20, mfc = 'c', mec = 'r')        #mec = Marker Edge Color
plt.show()
```

```python
plt.plot(ypoints, '--o', ms = 10, mec = '#000000', mfc = '#FF11FF')        
```

```python
plt.plot(ypoints, linestyle='dotted')
```

```python
plt.plot(ypoints, linestyle='dashed')
```

```python
plt.plot(ypoints, linestyle='-.')
plt.show()
```
