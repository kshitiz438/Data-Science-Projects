```python
#make a dic with new values to the chest pain
chest_pain_replacment_dic = {0 : "typical angina", 1 : "atypical angina", 2 : "non-anginal pain", 3 : "asymptomatic"}

df["cp"] = df["cp"].replace(chest_pain_replacment_dic)
```

```python
fasting_blood_sugar_replace = {0 : "less than or equal 120", 1 : "more than 120"}

df["fbs"]=df["fbs"].replace(fasting_blood_sugar_replace)
```

```python
elec_replacment = { 0 : "normal",
                  1 : "having ST-T wave abnormality",
                  2 : "left ventricular hyperthrophy"
                }
df["restecg"] = df["restecg"].replace(elec_replacment)
```

```python
thalassemia = { 1 : "downsloping",
              2 : "flat", 
              3 : "upsloping"}
df["thall"] = df["thall"].replace(thalassemia)
```

## Data Preprocessing

```python
# First we need to convert categorical variables into Numeric in order to find the most informative set of Feature
# to improve the performance of the classifier.

from sklearn.preprocessing import LabelEncoder
enc = {}
for col in df.select_dtypes(include = ['object']).columns:
    enc[col] = LabelEncoder()
    df[col] = enc[col].fit_transform(df[col])
    
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()

df[['trtbps']]=scaler.fit_transform(df[['trtbps']])
df[['chol']]=scaler.fit_transform(df[['chol']])
df[['thalachh']]=scaler.fit_transform(df[['thalachh']])
```

```python
df.head()
```

```python
# We have successfully converted Categorical values into numeric values
# Now we use a Heatmap to find the correlation between the features

corr = df.corr()
plt.figure(figsize = (20, 10))
sns.heatmap(corr, annot = True, linewidth = 0.5, cmap = 'YlGnBu')
plt.show()
```

```python
from sklearn.model_selection import train_test_split #Import train_test_split function
from sklearn import metrics #Import scikit-learn mertics function for precision, accuracy, confusion matrix calculations
X = df.drop('output', axis = 1) # Collecting the independent variables
y = df['output'] # Collecting the dependent Variable
```

```python
X.head()
```

```python
X.shape
```

```python
y.head()
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

## Decision Tree Classifier

```python
# Our First model is Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc = dtc.fit(X_train, y_train)
y_pred = dtc.predict(X_test)
```

```python
# Evaluating our Model

print("Accuracy: ", metrics.accuracy_score(y_test, y_pred))
print("Precision: ", metrics.precision_score(y_test, y_pred))
print("Recall: ", metrics.recall_score(y_test, y_pred))
print("F1 score: ", metrics.f1_score(y_test, y_pred))
```

```python
# Let's print the confusion matrix

cm = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4, 2))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', square=True, cbar=False, linewidth = 0.5)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
